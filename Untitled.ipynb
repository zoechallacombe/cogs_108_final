```python
import pandas as pd
import spotipy
import spotipy.util as util
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
#get raw data from github
topHundred = 'http://bit.ly/2V5uPUu'

#read csv
df_billboard = pd.read_csv(topHundred)
```

```python
#drop unnecessary columns
df_billboard.drop(['Instance'], axis=1, inplace = True)
```

```python
#convert cells in weekid column to dates
df_billboard['WeekID'] = pd.to_datetime(df_billboard['WeekID'])

#remove all rows with dates prior to 2018
#df_billboard = df_billboard[df_billboard['WeekID'].dt.year > 2012]

#remove additional rows with missing values 
df_billboard.dropna(inplace=True)

#resets the index of the dataframe
df_billboard.reset_index(drop=True,inplace=True)
```

```python
one_hit_billboard = df_billboard.drop_duplicates(subset=["Song"])
#one_hit_billboard = df_billboard.drop_duplicates(subset=["Performer"], keep=False)
```

```python
#for each song
dict_details = {}
for i in one_hit_billboard.iterrows():
    #get song and artist
    song = i[1]["Song"]
    artist=i[1]["Performer"]
    
    #search for song and artist in spotify with 5 results
    #TODO: will change this from just looking for song to looking for song and artist if not found?
    ###Well this didnt work :(
    query = song.split("(From", 1)[0]
    results = spotify.search(q=query, type='track', limit=20)
    print("SONG", song)
    print("ARTIST", artist)
    #print(results)
    track_id = ""
    missed = 0
    songs_not_found = []
    for j in range(20):
        #if the artist is in the key artist

        try:
            if (results['tracks']['items'][j]['artists'][0]['name'].lower() in artist.split("Featuring",1)[0].lower()):
                track_id = results['tracks']['items'][j]['id']
                break
        except IndexError as error:
            print("Song not found")
            
            track_id = "track id not found"
        

    track_details = spotify.audio_features(track_id)

    try:
        dict_details[track_id] = {key:val for key, val in track_details[0].items()}
    except AttributeError as error:
        dict_details[track_id] = {"": ""}

    dict_details[track_id]["song"] = song
    dict_details[track_id]["artist"] = artist.split("Featuring",1)[0]
    dict_details[track_id]["WeekID"] = i[1]["WeekID"]
```

```python
spotify_df = pd.DataFrame.from_dict(dict_details, orient='index')
```

```python
spotify.df = spotify_df.drop(spotify_df.columns[len(spotify_df.columns)-1], axis=1, inplace=True)
```

```python
spotify.df=spotify_df.drop(columns=["uri", "track_href","analysis_url", "id"], inplace=True)
```
